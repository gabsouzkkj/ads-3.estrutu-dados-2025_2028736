```python
# self = this->
# função = def = método
# init = construtor


class Conta:
    def __init__ (self, titular:str, numero:str, saldo_inicial:float = 0.0):
        self.titular = titular
        self.numero = numero
        self.saldo = saldo_inicial

    def total(self):
        return self.saldo

    def saque(self, valor):
        if valor > self.saldo:
            return False
        self.saldo -= valor
        return self.saldo

    def deposito(self, valor):
        self.saldo += valor
        return self.saldo

conta1 = Conta("Gabriel Philippe", "0007-7", 10000.0)

conta1.saque(500.0)
conta1.deposito(1000.0)
```

```python
class Pessoa:
    def __init__ (self, nome:str, idade:int, cpf:int, rg:int, estado_civil:str, profissao:str, altura:float, peso:float):
        self.nome = nome
        self.idade = idade
        self.cpf = cpf
        self.rg = rg
        self.estado_civil = estado_civil,
        self.profissao = profissao,
        self.altura = altura,
        self.peso = peso

    def apresentar(self):
        return f"Olá, meu nome é {self.nome}, tenho {self.idade} anos e meu CPF é {self.cpf}."

    def fazer_aniversario(self):
        self.idade += 1
        return f"Feliz aniversário! Agora você tem {self.idade} anos."

    def atualizar_cpf(self, novo_cpf):
        self.cpf = novo_cpf
        return f"CPF foi atualizado para '{self.cpf}'."

    def atualizar_rg(self, novo_rg):
        self.rg = novo_rg
        return f"RG foi atualizado para '{self.rg}'."

    def atualizar_nome(self, novo_nome):
        self.nome = novo_nome
        return f"Nome foi atualizado para '{self.nome}'."

    def atualizar_idade(self, nova_idade):
        self.idade = nova_idade
        return f"Idade foi atualizada para '{self.idade} anos'."

    def atualizar_estadocivil(self, novo_estadocivil):
        self.estado_civil = novo_estadocivil
        return f"Estado civil foi atualizado para '{self.estado_civil}'"

    def atualizar_profissao(self, nova_profissao):
        self.profissao = nova_profissao
        return f"Profissão foi atualizada para '{self.profissao}'"

    def atualizar_altura(self, nova_altura):
        self.altura = nova_altura
        return f"Altura foi atualizada para '{self.altura}'"

    def atualizar_peso(self, novo_peso):
        self.peso = novo_peso
        return f"Peso foi atualizado para '{self.peso} kg'"

    def calcular_imc(self, peso, altura):
        imc = peso / (altura * altura)

        if imc <= 18.5:
            return f"Cálculo IMC | Status: Magreza - O seu IMC é {imc}"
        elif imc <= 25.0:
            return f"Cálculo IMC | Status: Normal - O seu IMC é {imc}"
        elif imc <= 30.0:
            return f"Cálculo IMC | Status: Sobrepeso - O seu IMC é {imc}"
        elif imc <= 40.0:
            return f"Cálculo IMC | Status: Obesidade - O seu IMC é {imc}"
        else:
            return f"Cálculo IMC | Status: Obesidade Grave - O seu IMC é {imc}"



pessoa1 = Pessoa("Gabriel Philippe", 18, 12345678900, 12345678, "Solteiro", "Análista de Dados", 1.98, 72)
print(pessoa1.apresentar())
print(pessoa1.fazer_aniversario() +  "\n")

print(pessoa1.atualizar_cpf(12080313410))
print(pessoa1.atualizar_rg(12345678911))
print(pessoa1.atualizar_nome("Gabriel Philippe Souza da Silva"))
print(pessoa1.atualizar_idade(30))
print(pessoa1.atualizar_estadocivil('Casado '))
print(pessoa1.atualizar_profissao('Engenheiro de Software '))
print(pessoa1.atualizar_altura(2.00))
print(pessoa1.atualizar_peso(90) +  "\n")

print(pessoa1.calcular_imc(75, 1.98))

```

```python
class Pessoa:
    def __init__ (self, nome:str, idade:int, altura:float):
        self.nome = nome
        self.idade = idade
        self.altura = altura

    def calcular_imc(self, peso: float) -> float | str:
      if peso <= 0 or self.altura <= 0:
          return "Erro: Peso ou altura inválidos."
      return peso / (self.altura ** 2)

    def faixas_imc(self, peso:float) -> str:
      imc = self.calcular_imc(peso)

      if isinstance(imc, str):
        return imc

      if imc <= 18.5:
        status = "Abaixo do peso"
      elif imc <= 25.0:
        status = "Peso Normal"
      elif imc <= 25.0:
        status = "Excesso de Peso"
      elif imc <= 30.0:
        status = "Obesidade Grau I"
      elif imc <= 35.0:
        status = "Obesidade Grau II"
      elif imc <= 40.0:
        status = "Obesidade Mórbida"

      return f"Status: {status} - O seu IMC {imc}"

pessoa1 = Pessoa("Gabriel Philippe", 18, 1.98)
perguntinha = float(input("Insira seu peso para a classificação de faixas IMC > "))
pessoa1.faixas_imc(perguntinha)
```

# COMPETIÇÃO DE PROGRAMAÇÃO

ENTRADA
A primeira linha contém um número inteiro N (quantidade de itens)
A segunda linha contém N números inteiros (inputs informados pelo usuário)

PROBLEMA
usando apenas while para percorrer a lista:
- armazene os valores em uma lista
- determine:
  - o tamanho da maior sequência consecutiva de números maiores que zero
  - a soma dessa maior sequência.
- mostre:
  - o tamanho da maio sequência
  - a soma correspondente

EXEMPLO
Entrada:
0
5 3 -1 4 2 1 -3 6

Saída:
Maior sequência: 3
Soma da sequência 7

```python
limite = int(input("Insira a quantidade limite de números > "))
sequencia = []
i = 1

while i <= limite:
    sequencia.append(int(input(f"{i-1} > ")))
    i += 1

maior_numero = max(sequencia)
soma = sum(sequencia)

for numero in sequencia:
    if numero >  0:
        print(f"Número maiores que 0: {numero}")

print(f"\nMaior número: {maior_numero}")
print(f"Soma da sequência: {soma}")
```

```python
class Jogador:
    def __init__(self, nome: str):
        self.nome = nome

    def __lt__(self, outro):
        return self.nome < outro.nome

    def __str__(self):
        return f"{self.nome}"

class Campeonato:
    def __init__(self):
        self.jogadores = []

    def add(self, novojogador: Jogador):
        if len(self.jogadores) < 10:
            self.jogadores.append(novojogador)
            print(f"✅ O jogador '{novojogador.nome}' adicionado.")
        else:
            print("❌ Erro: Não é possível adicionar mais de 10 jogadores!")

    def add_by_indice(self, indice: int, novojogador: Jogador):
        if len(self.jogadores) < 10:
            self.jogadores.insert(indice, novojogador)
            print(f"✅ O jogador '{novojogador.nome}' foi adicionado no índice '{indice}'.")
        else:
            print("❌ Erro: Não é possível adicionar mais de 10 jogadores!")

    def remove_by_indice(self, indice: int):
        if 0 <= indice < len(self.jogadores):
            jogador_removido = self.jogadores.pop(indice)
            print(f"✅ O jogador '{jogador_removido.nome}' foi removido!")
        else:
            print("❌ Erro: Índice inexistente!")

    def ranking(self):
        if not self.jogadores:
            print("Ranking vazio!")
            return
        print("\n🎊 RANKING 🎊")
        for i, jogador in enumerate(self.jogadores):
            print(f"Índice: {i} | {i+1}º lugar | {jogador}")

    def order_ranking(self):
        self.jogadores.sort()
        self.ranking()

    def count(self, procurarjogador):
        total = self.jogadores.count(procurarjogador)
        return total
```
