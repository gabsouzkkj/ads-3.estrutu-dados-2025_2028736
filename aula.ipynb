```python
# self = this->
# função = def = método
# init = construtor


class Conta:
    def __init__ (self, titular:str, numero:str, saldo_inicial:float = 0.0):
        self.titular = titular
        self.numero = numero
        self.saldo = saldo_inicial

    def total(self):
        return self.saldo

    def saque(self, valor):
        if valor > self.saldo:
            return False
        self.saldo -= valor
        return self.saldo

    def deposito(self, valor):
        self.saldo += valor
        return self.saldo

conta1 = Conta("Gabriel Philippe", "0007-7", 10000.0)

conta1.saque(500.0)
conta1.deposito(1000.0)
```

```python
class Pessoa:
    def __init__ (self, nome:str, idade:int, cpf:int, rg:int, estado_civil:str, profissao:str, altura:float, peso:float):
        self.nome = nome
        self.idade = idade
        self.cpf = cpf
        self.rg = rg
        self.estado_civil = estado_civil,
        self.profissao = profissao,
        self.altura = altura,
        self.peso = peso

    def apresentar(self):
        return f"Olá, meu nome é {self.nome}, tenho {self.idade} anos e meu CPF é {self.cpf}."

    def fazer_aniversario(self):
        self.idade += 1
        return f"Feliz aniversário! Agora você tem {self.idade} anos."

    def atualizar_cpf(self, novo_cpf):
        self.cpf = novo_cpf
        return f"CPF foi atualizado para '{self.cpf}'."

    def atualizar_rg(self, novo_rg):
        self.rg = novo_rg
        return f"RG foi atualizado para '{self.rg}'."

    def atualizar_nome(self, novo_nome):
        self.nome = novo_nome
        return f"Nome foi atualizado para '{self.nome}'."

    def atualizar_idade(self, nova_idade):
        self.idade = nova_idade
        return f"Idade foi atualizada para '{self.idade} anos'."

    def atualizar_estadocivil(self, novo_estadocivil):
        self.estado_civil = novo_estadocivil
        return f"Estado civil foi atualizado para '{self.estado_civil}'"

    def atualizar_profissao(self, nova_profissao):
        self.profissao = nova_profissao
        return f"Profissão foi atualizada para '{self.profissao}'"

    def atualizar_altura(self, nova_altura):
        self.altura = nova_altura
        return f"Altura foi atualizada para '{self.altura}'"

    def atualizar_peso(self, novo_peso):
        self.peso = novo_peso
        return f"Peso foi atualizado para '{self.peso} kg'"

    def calcular_imc(self, peso, altura):
        imc = peso / (altura * altura)

        if imc <= 18.5:
            return f"Cálculo IMC | Status: Magreza - O seu IMC é {imc}"
        elif imc <= 25.0:
            return f"Cálculo IMC | Status: Normal - O seu IMC é {imc}"
        elif imc <= 30.0:
            return f"Cálculo IMC | Status: Sobrepeso - O seu IMC é {imc}"
        elif imc <= 40.0:
            return f"Cálculo IMC | Status: Obesidade - O seu IMC é {imc}"
        else:
            return f"Cálculo IMC | Status: Obesidade Grave - O seu IMC é {imc}"



pessoa1 = Pessoa("Gabriel Philippe", 18, 12345678900, 12345678, "Solteiro", "Análista de Dados", 1.98, 72)
print(pessoa1.apresentar())
print(pessoa1.fazer_aniversario() +  "\n")

print(pessoa1.atualizar_cpf(12080313410))
print(pessoa1.atualizar_rg(12345678911))
print(pessoa1.atualizar_nome("Gabriel Philippe Souza da Silva"))
print(pessoa1.atualizar_idade(30))
print(pessoa1.atualizar_estadocivil('Casado '))
print(pessoa1.atualizar_profissao('Engenheiro de Software '))
print(pessoa1.atualizar_altura(2.00))
print(pessoa1.atualizar_peso(90) +  "\n")

print(pessoa1.calcular_imc(75, 1.98))

```
